Using a Monte Carlo technique, calculate the area of a circle with radius=5.2

Show visualization, i.e. plot, of your method

```python
import numpy as np
import matplotlib.pyplot as plt

circle_eq = lambda x,y: np.sqrt(x**2 + y**2)

r = 5.2
n_samples = 10000
# sample x and y:
x_sample = np.random.uniform(0,5.2, n_samples)
y_sample = np.random.uniform(0,5.2, n_samples)

pred_rs = circle_eq(x_sample, y_sample)

kept_values = np.argwhere(pred_rs <= r)
reject_values = np.argwhere(pred_rs > r)

# find the ratio of the circles: 
total_area = r**2 #len(n_samples)

# because A_circle = pi*r**2, we cen get pi from the ratio between total number samples / total samples in the circle. 

pi_estimate = 4*len(kept_values)/n_samples
print("predicted pi value: ", pi_estimate)

plt.scatter(x_sample[kept_values], y_sample[kept_values])
plt.scatter(x_sample[reject_values], y_sample[reject_values]) 
plt.show()

# to calcualate the area: 
# we can use the ratio of the circle area to the square area to calculate pi.
total_area = r**2 # proportional to the total number of samples
#circle_area (unknown) # proportional to the number of samples in the circle
circle_area = 4*total_area * len(kept_values)/n_samples
print("circle area: ", circle_area)
print("true values of cicle area: " , np.pi*r**2)



```
