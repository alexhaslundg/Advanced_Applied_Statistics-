# Uncertainties in Areas 


```python
import numpy as np 
import matplotlib.pyplot as plt
import math


def volume_sphere(n, r):
    """Compute the volume of an n-dimensional sphere with radius r."""
    return (math.pi ** (n / 2) / math.gamma(n / 2 + 1)) * (r ** n)

def volume_ellipsoid(n, axes):
    """Compute the volume of an n-dimensional ellipsoid given its semi-axes."""
    if len(axes) != n:
        raise ValueError("Number of axes must match the dimension n")
    return (math.pi ** (n / 2) / math.gamma(n / 2 + 1)) * math.prod(axes)

# Example usage:
n = 3  # Dimension
r = 1  # Radius of sphere
print("Sphere Volume:", volume_sphere(n, r))



r = 10 # define the radius of the sphere
dim_range = range(1, 100)
v_spheres = np.array([volume_sphere(n, r) for n in dim_range])
v_cube = np.array([(2*r)**n for n in dim_range])

plt.plot(dim_range, v_spheres, 'o-', color='blue', label='Sphere')
plt.plot(dim_range, v_cube, 'o-', color='red', label='Cube')
plt.legend()
plt.xlabel("Dimension")
plt.ylabel("Volume")
plt.yscale('log')
plt.show()


percent_error = 100*(v_cube-v_spheres) / v_spheres
plt.plot(dim_range,percent_error , 'o-', color='blue')
plt.xlabel("Dimension")
plt.ylabel("Percent Error In Volume")
plt.title("Volume Ratio of Sphere to Cube")
plt.yscale('log')
plt.show()

# find were the cube and sphere volumes are equal
dim = np.argmin(np.abs(v_spheres - v_cube))
print("The sphere and cube have the same volume in dimension", dim+1)

plt.plot(dim_range, v_spheres-v_cube, 'o-', color='blue', label='Sphere')
plt.yscale('log')
```
